```python
import sqlite3
import pandas as pd
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table'"
```

```python
con = sqlite3.connect("/content/nba_salary.sqlite")
df_tables = pd.read_sql_query(query,con)
print(df_tables)

con.close()
```

```python
# Ler o dataset utilizado para extrair apenas os jogadores com um salário maior que 10 milhões de dólares

value = 10_000_00

query = f'''
SELECT *
FROM NBA_season1718_salary
WHERE season17_18 > {value}
ORDER BY season17_18;
'''

con = sqlite3.connect("/content/nba_salary.sqlite")
df_salaries_greater_than_10000000 = pd.read_sql_query(query, con)
df_salaries_greater_than_10000000.head()
```

```python
# Extrair as estatísticas desses jogadores correspondentes ao ano fornecido na tabela Season_Stats do ano 2017
year = 2017

query = f'''
SELECT *
FROM Seasons_Stats
WHERE Year = {year}
ORDER BY Year;
'''

con = sqlite3.connect("/content/nba_salary.sqlite")
df_stats_2017 = pd.read_sql_query(query, con)
df_stats_2017.head()
```

```python
# Calcular a média total de pontos PTS, assistências AST e faltas pessoais PF da tabela Season_Stats.

query = f'''
SELECT AVG(PTS) AS PTS_AVG,
       AVG(AST) AS AST_AVG,
       AVG(PF) AS PF_AVG
FROM Seasons_Stats;
'''

con = sqlite3.connect("/content/nba_salary.sqlite")
df_stats_mean_PTS_AST_PF = pd.read_sql_query(query, con)
df_stats_mean_PTS_AST_PF.head()
```
