```python
# entendendo e exemplo de list comprehension
lista_colunas_com_nu = [coluna for coluna in df_enem_2022.columns if 'NU' in coluna]
lista_colunas_com_nu
```

```python
# capturando as colunas com as respostas para removê-las, já que não vamos analisar no nível de questões das provas
colunas_questoes = [coluna for coluna in df_enem_2022.columns if coluna.startswith('TX_')]
colunas_questoes
```

```python
# as colunas do tipo CO_ não serão utilizadas na análise, são só códigos de coisas que temos siglas (UF, por exemplo)
colunas_codigos = [coluna for coluna in df_enem_2022.columns if coluna.startswith('CO_')]
colunas_codigos
```

```python
# não vamos fazer também uma análise por municípios, então vamos remover tudo que tiver essa string
colunas_municipios = [coluna for coluna in df_enem_2022.columns if 'MUNICIPIO' in coluna]
colunas_municipios
```

```python
# removendo as colunas (inscrição e ano não são necessários)
colunas_remover = colunas_questoes + colunas_codigos + colunas_municipios + ['NU_INSCRICAO', 'NU_ANO']
df_enem_2022 = df_enem_2022.drop(colunas_remover, axis=1)
df_enem_2022.head()
```

```python
from string import ascii_uppercase

dicionario_acesso_internet = dict()

# iterando sobre o alfabeto (python não é lindo?)
for i, letra in enumerate(ascii_uppercase):
    if i == 2:
        break
    dicionario_acesso_internet[letra] = i
dicionario_acesso_internet
```

```python
# analisando as notas por acesso à internet (sim ou não)

# Configurações do Seaborn
sns.set_style("darkgrid")

# Criar a figura e os subplots
figura, graficos = plt.subplots(1, 4, sharey=True, figsize=(8,6))
plt.tight_layout()

# Mapeamento de 0 para 'N' e 1 para 'S'
mapeamento_acesso = {0: 'N', 1: 'S'}

# Aplicar o mapeamento para a coluna 'acesso_a_internet'
df_acesso_a_internet['acesso_a_internet'] = df_acesso_a_internet['acesso_a_internet'].map(mapeamento_acesso)

# Criar os gráficos boxplot
for i in range(4):
    coluna_1 = colunas_notas[i]
    sns.boxplot(data=df_acesso_a_internet, y=coluna_1, x='acesso_a_internet', ax=graficos[i])

# Mostrar o gráfico
plt.show()
```
